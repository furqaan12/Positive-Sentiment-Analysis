```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import defaultdict

# join the text from the DataFrame
text1 = " ".join(text for text in df.text)

# generate the word cloud
word_cloud1 = WordCloud(collocations = False, background_color = 'white',
                        width = 2048, height = 1080).generate(text1)

# extract the words and their frequencies from the wordcloud object
word_freq = defaultdict(int)
for word, freq in word_cloud1.words_.items():
    word_freq[word] = freq

# print out the words and their frequencies
for word, freq in word_freq.items():
    print(f"Word: {word}, Intensity: {freq}")

# save and show the image
word_cloud1.to_file('got.png')
plt.imshow(word_cloud1, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import csv

# write the data to a CSV file
with open('word_frequency.csv', 'w', newline='') as csvfile:
    fieldnames = ['Word', 'Intensity']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for word, freq in word_freq.items():
        writer.writerow({'Word': word, 'Intensity': freq})
```
